```python
import pandas as pd
import os

# Set full paths
base_dir = r"C:\Users\Admin\Desktop\INX_Employee_Performance_Project"
input_file = os.path.join(base_dir, "data", "raw", "INX_Employee_Performance.xlsx")
output_dir = os.path.join(base_dir, "data", "processed")

# ✅ Ensure the processed directory exists
os.makedirs(output_dir, exist_ok=True)

# Load data
df = pd.read_excel(input_file)
print("✅ Raw data loaded successfully.")
```

```python
# Clean data
# Drop columns with more than 40% missing values
threshold = 0.4
cols_to_drop = [col for col in df.columns if df[col].isnull().mean() > threshold]
df_cleaned = df.drop(columns=cols_to_drop)

# Fill remaining missing values
df_cleaned = df_cleaned.ffill()

# Save cleaned data
output_file = os.path.join(output_dir, "cleaned_employee_data.csv")
df_cleaned.to_csv(output_file, index=False)

print(f"✅ Cleaned data saved at: {output_file}")
```
